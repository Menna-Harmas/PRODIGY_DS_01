```python
import random
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("C:\\Dell\\prodigy_task_1\\world_population.csv")
df.head()
```

```python
final_data = df.drop(['Rank','CCA3','Capital','Continent','Area (km²)','Density (per km²)','World Population Percentage'], axis=1)
final_data.head()
```

```python
countries = ['India','Nigeria','Pakistan','China','Algeria']

countries_data = final_data[final_data['Country/Territory'].isin(countries)]
countries_data
```

```python
data = {
    'India': list(countries_data[countries_data['Country/Territory'] == 'India'].iloc[:,1:9].values.flatten()),
    'Nigeria': list(countries_data[countries_data['Country/Territory'] == 'Nigeria'].iloc[:,1:9].values.flatten()),
    'Pakistan': list(countries_data[countries_data['Country/Territory'] == 'Pakistan'].iloc[:,1:9].values.flatten()),
    'China': list(countries_data[countries_data['Country/Territory'] == 'China'].iloc[:,1:9].values.flatten()),
    'Algeria': list(countries_data[countries_data['Country/Territory'] == 'Algeria'].iloc[:,1:9].values.flatten())
}

years = list(countries_data.columns[1:9])
years = [year.replace(" Population",'') for year in years]
```

```python
plt.figure(figsize=(15, 8))

for country, population in data.items():
    plt.plot(years, population, label=country)

plt.annotate('India expected to become\nthe most populous\n country by 2030', xy=(years[-2], data['India'][-2]),
             xytext=(years[-2], data['India'][-2] + 100000000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=11, ha='center')
plt.annotate('Expected population\n decline\nin China', xy=(years[0], data['China'][0]),
             xytext=(years[0], data['China'][0] + 100000000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=11, ha='center')
plt.annotate('Population doubling\n expected in\nsome African\n countries', xy=(years[-1], data['Nigeria'][-1]),
             xytext=(years[-1], data['Nigeria'][-2] + 100000000),
             arrowprops=dict(facecolor='black', arrowstyle='->'),
             fontsize=11, ha='center')  


plt.xlabel('Years')  
plt.ylabel('Population')
plt.title('Population Growth Over the Years', fontsize=16)
plt.legend()

plt.xticks(rotation=45)
plt.style.use('dark_background')
plt.show()
```

```python

chosen_country = random.choice(countries)

years = list(countries_data.columns[1:9])
years = [year.replace(" Population",'') for year in years]

population = list(countries_data[countries_data['Country/Territory'] == chosen_country].iloc[:,1:9].values.flatten())

plt.figure(figsize=(10, 6))
plt.bar(years, population, color='skyblue',alpha=.9)
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')
plt.xticks(rotation=45)  
plt.tight_layout()  
plt.style.use('dark_background')
plt.show()
```
